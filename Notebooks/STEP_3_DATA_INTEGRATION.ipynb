### - LET US LOAD OUR GEOJSON FILE INTO THIS NOTEBOOK USING GEOPANDAS

```python
## Here we imported the python pandas and geoandas packages needed for working with the geojson file
import pandas as pd
import geopandas as gpd
```

```python
##here we read the geojson file into the notebook as a dataframe for easy manipulations
geojson = gpd.read_file("../Data/Nigeria_Poly.geojson")
```

### SCRAPING OF DATA FROM THE NCDC SITE USING ".read_html"

```python
data = pd.read_html("https://covid19.ncdc.gov.ng/")
```

```python
data
```

```python
##Selecting the table we need from the scraped data
ncdc_data = data[-2].copy()
```

```python
ncdc_data
```

### - LETS PLAY WITH THIS DATAFRAMES AND INTEGRATE THEM TOGETHER INTO A SINGLE GEOJSON FILE WHICH WILL BE USED FOR THE FOLIUM MAP GENERATION

```python
##This function gives both dataframes a unique name for Abuja...from above it can be seen that both dataframes have differrent names referring to same location
def name_fct (df):
    if df in ["Abuja FCT", "Federal Capital Territory"]:
        new_df = "FCT"
        return new_df
    else:
        return df
```

```python
##Applying the functions to the DataFrames
geojson["Name"] = geojson["Name"].apply(name_fct)
ncdc_data["States Affected"] = ncdc_data["States Affected"].apply(name_fct)
```

```python
geojson
```

```python
ncdc_data
```

```python
## The functions below help us to extract comfirmed cases, active cases, no of discahrges and no of death form the ncdc dataframe and map it to the geojson dataframe

def get_confirmed_cases (df):
    s = ncdc_data["States Affected"]
    if df in list(s):
        i = list(s).index(df)
        return ncdc_data.loc[i, "No. of Cases (Lab Confirmed)"]
    else:
        return 0
    
def get_active_cases (df):
    s = ncdc_data["States Affected"]
    if df in list(s):
        i = list(s).index(df)
        return ncdc_data.loc[i, "No. of Active Cases"]
    else:
        return 0
    
def get_discharged (df):
    s = ncdc_data["States Affected"]
    if df in list(s):
        i = list(s).index(df)
        return ncdc_data.loc[i, "No. Discharged"]
    else:
        return 0

    
def get_death (df):
    s = ncdc_data["States Affected"]
    if df in list(s):
        i = list(s).index(df)
        return ncdc_data.loc[i, "No of Deaths"]
    else:
        return 0


```

```python
geojson["Confirmed Cases"] = geojson["Name"].apply(get_confirmed_cases)
geojson["Active Cases"] = geojson["Name"].apply(get_active_cases)
geojson["No. Discharged"] = geojson["Name"].apply(get_discharged)
geojson["Deaths"] = geojson["Name"].apply(get_death)
```
